Inactive or Resigned Employees

```python
import pandas as pd
from datetime import datetime
```

```python
df = pd.read_csv('../00_Data/01_LatestRecords.csv')
df.head()
```

```python
df['Joining'].dtype, df['Termination'].dtype
```

```python
df['Joining'] = pd.to_datetime(df['Joining'])
df['Termination'] = pd.to_datetime(df['Termination'])
```

```python
df.dtypes
```

```python
def calculate_active_months(start_date:datetime, end_date:datetime):
    """CALCUALTE ACTIVE MONTHS"""
    months = (end_date.year - start_date.year)*12 + (end_date.month - start_date.month)
    return months
```

```python
df_inactive = df.copy()
df_inactive.loc[:, 'active_months'] = df_inactive.apply(lambda x: 
                                                        calculate_active_months(x['Joining'], 
                                                                                x['Termination']), axis=1)
df_inactive.head()                                                                         
```

```python
df_inactive = df_inactive.loc[df['Salary Status'] == 'Inactive']
```

```python
df_inactive.to_csv('../00_Data/02_ResignedEmployees.csv', index=False)
```
