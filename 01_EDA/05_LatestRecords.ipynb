Optimizing Rows | Considering Latest Records only

```python
import pandas as pd
```

```python
df = pd.read_csv('../00_Data/data.csv')
df.head()
```

```python
df.shape
```

```python
df.loc[df['Employee ID'] == 100001]
```

```python
df['Salary Date'] = pd.to_datetime(df['Salary Date'])
df.dtypes
```

```python
df_latest = df.loc[df['Salary Date'] == df.groupby('Employee ID')['Salary Date'].transform('max')]
df_latest
```

```python
df_latest.shape
```

```python
df_latest.to_csv('../00_Data/01_LatestRecords.csv', index=False)
```
